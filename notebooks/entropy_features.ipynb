```python
import pandas as pd

from weschatbot.ambiguity.entropy import discrete_softmax_entropy_cosine
```

```python
file_name = "messages_202512292119_chunks"
```

```python
df = pd.read_parquet(f"../data/{file_name}.parquet")
```

```python
results = []
for qid, group in df.groupby("question_id"):
    scores = group["score"].tolist()
    entropy_info = discrete_softmax_entropy_cosine(scores, t=0.05)

    results.append({
        "question_id": qid,
        "entropy": entropy_info["entropy"],
        "normalized_entropy": entropy_info["normalized_entropy"],
        "probs": entropy_info["probs"].tolist()
    })

entropy_df = pd.DataFrame(results)

entropy_df.to_parquet(f"../data/{file_name}_entropy_results.parquet", engine="pyarrow", index=False)

print(entropy_df)
```
