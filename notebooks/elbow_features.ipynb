```python
import pandas as pd

from weschatbot.ambiguity.elbow_detection import gaussian_weighted_elbow
from weschatbot.ambiguity.steepness import steepness_top1_elbow
```

```python
file_name = "messages_202512292119_chunks"
```

```python
df = pd.read_parquet(f"../data/{file_name}.parquet")

```

```python
elbow_func = gaussian_weighted_elbow(alpha=0.5, min_index=1, sigma_factor=0.2)

results = []
for qid, group in df.groupby("question_id"):
    scores = group["score"].tolist()
    idx, val = elbow_func(scores)

    steepness_info = steepness_top1_elbow(scores, elbow_idx=idx)

    results.append({
        "question_id": qid,
        "elbow_index": idx,
        "elbow_value": val,
        "gap": steepness_info["gap"],
        "gap_norm": steepness_info["gap_norm"],
        "pos_weight": steepness_info["pos_weight"],
        "combined": steepness_info["combined"]
    })

elbow_df = pd.DataFrame(results)
print(elbow_df)
```

```python
elbow_df.to_parquet(f"../data/{file_name}_elbow_steepness_results.parquet", engine="pyarrow", index=False)
```
